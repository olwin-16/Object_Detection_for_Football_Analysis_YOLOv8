# **Training YOLO8 for Object Detection**

* **task=** detect or segment
* **mode=** train, val
* **model=** path to model file, i.e. yolov8n.pt, yolov8n.yaml
* **data=** path to data file dataset.yaml
* **epochs=** number of epochs to train for
* **imgsz=** size of input images as integer
* **batch=** number of images per batch
* **project=** project traing results saving path
* **name=** experiment name


```python
import os

train_images = os.listdir('/content/drive/MyDrive/Colab Notebooks/ObjectDetection/Dataset/train/images')
print("Train Images:", train_images)

val_images = os.listdir('/content/drive/MyDrive/Colab Notebooks/ObjectDetection/Dataset/valid/images')
print("Validation Images:", val_images)
```
